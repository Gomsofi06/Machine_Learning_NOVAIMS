*Input `Age at Injury` and variables that can be calculated with it*

```python
median_age_at_injury = X_train['Age at Injury'].median()

X_train['Age at Injury'] = X_train['Age at Injury'].fillna(median_age_at_injury)
X_val['Age at Injury'] = X_val['Age at Injury'].fillna(median_age_at_injury)
X_test['Age at Injury'] = X_test['Age at Injury'].fillna(median_age_at_injury)
```

```python
def calculate_birth_year(df):
    # Ensure the correct format of 'Birth Year'
    df['Accident Date'] = pd.to_datetime(df['Accident Date'], errors='coerce')

    # Filter the rows where 'Birth Year' is NaN, but 'Age at Injury' and 'Accident Date' are not NaN
    condition = df['Birth Year'].isna() & df['Age at Injury'].notna() & df['Accident Date'].notna()

    # Replace missing 'Birth Year' with the difference between 'Accident Date' year and 'Age at Injury'
    df.loc[condition, 'Birth Year'] = df.loc[condition, 'Accident Date'].dt.year - df.loc[condition, 'Age at Injury']
```

```python
calculate_birth_year(X_train)
calculate_birth_year(X_val)
calculate_birth_year(X_test)
```

Scale the data:

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)
```

## <span style="color:salmon"> 3. Neural Networking </span>

NOTA: Hiperpar√¢metros Recomendados

- hidden_layer_sizes (Arquitetura da rede):
    - Redes profundas geralmente t√™m 2-4 camadas escondidas.
    - N√∫mero de neur√¥nios em cada camada deve ser ajustado com base na complexidade do dataset.
    - Sugest√£o inicial: testar configura√ß√µes como (64, 32), (128, 64, 32) ou (100, 50).
- activation (Fun√ß√£o de ativa√ß√£o):
    - Para dados complexos:
    - relu (ReLU): Melhor para grandes redes e tarefas gerais.
    - tanh: Pode ser √∫til para dados com distribui√ß√£o centrada em torno de 0.
    - Evite logistic (sigm√≥ide) em redes profundas, pois sofre com problemas de gradiente desaparecido.
- solver (Otimizador):
    - adam: R√°pido e eficiente para a maioria das situa√ß√µes.
    - lbfgs: Melhor para datasets pequenos ou quando o treinamento converge rapidamente.
    - sgd: √ötil se ajustar uma taxa de aprendizado manual.
- alpha (Regulariza√ß√£o L2):
    - Controla overfitting. Teste valores entre 0.0001 e 0.01.
    - Sugest√£o: [0.0001, 0.001, 0.01].
- max_iter (N√∫mero m√°ximo de itera√ß√µes):
    - Valores como 200-500 s√£o suficientes para a maioria dos datasets.
    - Ajuste se a rede n√£o convergir.

Applies MLP classifier:

```python
model_mlp = MLPClassifier(hidden_layer_sizes=(64, 32), activation='relu', solver='adam', max_iter=200, random_state=42).fit(X_train, y_train)
```

```python
y_train_pred = model_mlp.predict(X_train)
```

```python
y_val_pred = model_mlp.predict(X_val)
```

#### <span style="color:salmon"> 3.1  Evaluate the model </span> 


```python
class_mapping = {
    0:'1. CANCELLED', 
    1:'2. NON-COMP',
    2:'3. MED ONLY', 
    3:'4. TEMPORARY',
    4:'5. PPD SCH LOSS', 
    5:'6. PPD NSL', 
    6:'7. PTD', 
    7:'8. DEATH'
}

# Use the values from class_mapping as the target names
target_names = list(class_mapping.values())
```

Compute confusion matrix to evaluate the accuracy of a classification

```python
# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_train, y_train_pred))
print("\nClassification Report:")
print(classification_report(y_train, y_train_pred, target_names=target_names))
```

```python
# Evaluate the model
print("Confusion Matrix:")
print(confusion_matrix(y_val, y_val_pred))
print("\nClassification Report:")
print(classification_report(y_val, y_val_pred, target_names=target_names))
```
