## Metadata

**1.**  *Accident Date*: Injury date of the claim.  

**2.**  *Age at Injury*: Age of injured worker when the injury occurred.  

**3.**  *Alternative Dispute Resolution*: Adjudication processes external to the Board.  

**4.**  *Assembly Date*: The date the claim was first assembled.  

**5.**  *Attorney/Representative*: Is the claim being represented by an Attorney?  

**6.**  *Average Weekly Wage*: The wage used to calculate workers’ compensation, disability, or Paid Leave wage replacement benefits.  

**7.**  *Birth Year*: The reported year of birth of the injured worker.  

**8.**  *C-2 Date*: Date of receipt of the Employer's Report of Work-Related Injury/Illness or equivalent (formerly Form C-2).  

**9.**  *C-3 Date*: Date Form C-3 (Employee Claim Form) was received.  

**10.** *Carrier Name*: Name of primary insurance provider responsible for providing workers’ compensation coverage to the injured worker's employer.  

**11.** *Carrier Type*: Type of primary insurance provider responsible for providing workers’ compensation coverage.  

**12.** *Claim Identifier*: Unique identifier for each claim, assigned by WCB.  

**13.** *County of Injury*: Name of the New York County where the injury occurred.  

**14.** *COVID-19 Indicator*: Indication that the claim may be associated with COVID-19.  

**15.** *District Name*: Name of the WCB district office that oversees claims for that region or area of the state.  

**16.** *First Hearing Date*: Date the first hearing was held on a claim at a WCB hearing location. A blank date means the claim has not yet had a hearing held.  

**17.** *Gender*: The reported gender of the injured worker.  

**18.** *IME-4 Count*: Number of IME-4 forms received per claim.  

**19.** *Industry Code*: NAICS code and descriptions used to classify businesses according to their economic activity.  

**20.** *Industry Code Description*: 2-digit NAICS industry code description used to classify businesses according to their economic activity.  

**21.** *Medical Fee Region*: Approximate region where the injured worker would receive medical service.  

**22.** *OIICS Nature of Injury Description*: The OIICS nature of injury codes and descriptions are available online.  

**23.** *WCIO Cause of Injury Code*: The WCIO cause of injury codes and descriptions are available online.  

**24.** *WCIO Cause of Injury Description*: See the description above for WCIO Cause of Injury Code.  

**25.** *WCIO Nature of Injury Code*: The WCIO nature of injury codes are available online.  

**26.** *WCIO Nature of Injury Description*: See the description above for WCIO Nature of Injury Code.  

**27.** *WCIO Part Of Body Code*: The WCIO part of body codes and descriptions are available online.  

**28.** *WCIO Part Of Body Description*: See the description above for WCIO Part of Body Code.  

**29.** *Zip Code*: The reported ZIP code of the injured worker’s home address.  

<u>(Binary)</u> **30.** *Agreement Reached*: 'Yes' if there is an agreement without the involvement of the WCB; unknown at the start of a claim.  

<u>(Multiclass)</u> **31.** *WCB Decision*: ecision of the WCB relative to the claim.  

<u>(Target)</u> **32.** *Claim Injury Type*: Deliberation of the WCB relative to benefits awarded to the claim.  

## Concepts

**1.**  *C-2 Form*: Employer's Report of Work-Related Injury/Illness.  

**2.**  *C-3 Form*: Employee Claim Form for workers' compensation.  

**3.**  *IME*: Independent Medical Examination.  

**4.**  *OIICS*: Occupational Injury and Illness Classification System.  

**5.**  *WCIO*: Workers' Compensation Insurance Organizations.  

**6.**  *NAICS*: North American Industry Classification System.  

**7.**  *WCB*: Workers' Compensation Board.  

**8.**  *ADR*: Alternative Dispute Resolution.

```python
convert_date = ['Accident Date', 'Assembly Date', 'C-2 Date', 'C-3 Date', 'First Hearing Date']

# Loop through the columns
for col in convert_date:
    # Convert the columns with errors coerced to NaN, then cast to datetime64[ns]
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Check if the conversion was properly done
print(df[convert_date].dtypes)
```

```python
convert_int = ['Age at Injury', 'Birth Year', 'IME Count']

# Loop through the columns
for col in convert_int:
    # Convert the columns with errors coerced to NaN, then cast to nullable Int64
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')

# Check if the conversion was properly done
print(df[convert_int].dtypes)
```

```python
convert_float = ['']

# Loop through the columns
for col in convert_float:
    # Convert the columns with errors coerced to NaN, then cast to float64
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('float64')

# Check if the conversion was properly done
print(df[convert_float].dtypes)
```

Handling outliers (approach suggestion)

```python
# Removing Outliers
# using IQR method

for column in num_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```
